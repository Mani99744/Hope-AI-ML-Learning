```python
# Import pandas library for data manipulation and analysis
import pandas as pd
```

```python
# Read the salary_data dataset from a CSV file into a pandas DataFrame
dataset = pd.read_csv("salary_data.csv")
```

```python
# Extract 'YearsExperience' column from dataset and store as independent variable
independent = dataset[["YearsExperience"]]
```

```python
# Extract 'Salary' column from dataset and store it as dependent variable
dependent = dataset[['Salary']]
```

```python
# Import train_test_split function from scikit-learn's model_selection module
from sklearn.model_selection import train_test_split
```

```python
# Split dataset into training (70%) and testing (30%) sets using train_test_split
# independent: Features/predictors (X variables)
# dependent: Target variable (Y variable)
# test_size=0.30: 30% of data used for testing
# random_state=0: Set seed for reproducibility
X_train,X_test,Y_train,Y_test = train_test_split(independent, dependent , test_size = 0.30, random_state = 0)
```

```python
# Import the LinearRegression class from scikit-learn's linear_model module
from sklearn.linear_model import LinearRegression

# Create an instance of LinearRegression model for performing linear regression
regressor = LinearRegression()
```

```python
# Train the regression model using the training data (X_train for features, Y_train for target values)
regressor.fit(X_train,Y_train)
```

```python
# Extract the coefficients (weights)
weight = regressor.coef_
```

```python
# Access the 'weight' variable to retrieve or display weight values
weight
```

```python
# Get the bias (intercept) term 
bias = regressor.intercept_
```

```python
# Use the trained regressor model to make predictions on the test set
Y_pred = regressor.predict(X_test)
```

```python
# Import r2_score from scikit-learn metrics module to calculate the coefficient of determination (RÂ²)
from sklearn.metrics import r2_score
```

```python
# Calculate R-squared score to evaluate model performance by comparing test values with predictions
r_score = r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
# Import pickle module to enable serialization and deserialization of Python objects
import pickle
```

```python
# Define the filename for saving the model
filename = "Finalised_model_linear.sav"
# Save the trained linear regression model to a file using pickle
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Load the previously saved machine learning model from a file using pickle
# 'rb' flag opens the file in binary read mode
loaded_model = pickle.load(open(filename,'rb'))
```

```python
# Use the trained model to predict the output for input value 48
result = loaded_model.predict([[48]])
```
